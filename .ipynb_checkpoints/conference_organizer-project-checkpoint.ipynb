For each partner, you need to categorize them according to the country in which they must attend the meeting. Find a way to determine which availability dates work best for everyone in EACH country. If a date doesn't work for a specific partner, they simply will not be able to attend the meeting. 

<strong>The kicker</strong>: this meeting will last a total of two days. Because of this, you must find a RANGE of two dates that work best for everyone. For whichever two days work the best, only display the first date everyone can start.

```python
# Pulling API Data
import requests

page_link = 'https://ct-mock-tech-assessment.herokuapp.com/'

page = requests.get(page_link)

data = page.json()

#232 partners
guests = data['partners']

#create a list after for loop of attributes you want vs data attributes
guest_list = [] 

for i in guests:
#     print(i)
    new_dict = {
        'availability': i['availableDates'],
        'location': i['country'],
        'contact': i['email'],
        'f_name': i['firstName'],
        'l_name': i['lastName'],
    }
    guest_list.append(new_dict)
#     print(new_dict)
#     break

guest_list

# print out api data in order 
class Guest:#*args
    def __init__(self, *args):
        self.args = args

# attributes created
    def from_dict(self, data):
        for field in ['availability', 'location', 'contact', 'f_name', 'l_name']:
            if field in data:
                setattr(self, field, data[field])
                
# this turns data into human form                    
    def __repr__(self):
        return f'<Guest: {self.f_name} {self.l_name} {self.location} {self.contact} \n{self.availability}>'

# connects it all into a organized list
new_guest_list = []

for guest_dict in guest_list:
    new_guest_obj = Guest()
    new_guest_obj.from_dict(guest_dict)
    new_guest_list.append(new_guest_obj)
    
new_guest_list

```
